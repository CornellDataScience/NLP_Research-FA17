```python
import pickle
with open("review_user_top10_5star_list.txt", "rb") as f:
    review_user_top10_5star_list = pickle.load(f)
```

```python
import pickle
with open("review_user_top10_1star_list.txt", "rb") as f:
    review_user_top10_1star_list = pickle.load(f)
```

```python
import pickle
with open("review_user_top10_list.txt", "rb") as f:
    review_user_top10_list = pickle.load(f)
```

```python

import nltk
 
def format_sentence(sent):
    return({word: True for word in nltk.word_tokenize(sent)})
 
print(format_sentence("The cat is very cute"))
```

```python
neg = []

for i in review_user_top10_1star_list: 
    neg.append([format_sentence(i), 'neg'])
```

```python
neg[:2]
```

```python
pos = []

for i in review_user_top10_5star_list: 
    pos.append([format_sentence(i), 'pos'])
```

```python
training = pos[:int((.8)*len(pos))] + neg[:int((.8)*len(neg))]
```

```python

from nltk.classify import NaiveBayesClassifier
 
classifier = NaiveBayesClassifier.train(training)
```

```python
classifier.show_most_informative_features()
```

```python
res = []
for i in review_user_top10_list:
    res.append([i, classifier.classify(format_sentence(i))])
    print(classifier.classify(format_sentence(i)))
```
