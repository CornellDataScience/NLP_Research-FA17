### Apply Haversine formula to calculate distance between user and restaurant

```python
from haversine import haversine
def calculate_distance(row):
    loc1 = (row['latitude'], row['longitude'])
    loc2 = (row['gmm_latitude'], row['gmm_longitude'])
    return haversine(loc1, loc2)
```

```python
date['distance'] = date.apply(calculate_distance, axis=1)
```

```python
date
```

### Plot of reviews by locality (distance from restaurant)
Distance measured in Kilometers

```python
import matplotlib.pyplot as plt
date['distance'].plot(kind='hist')
plt.show()
```

```python
def time_span(user):
    user = [user]
    tmp = date.query('user_id == @user')['date']
    late = tmp.max()
    early = tmp.min()
    return abs((late - early).days)
```

```python
user = date['user_id'].drop_duplicates().reset_index()
del user['index']
user['time_span']=pd.Series()
user['time_sd']=pd.Series()
user
```

```python
for i in range(4000):
    user['time_span'][i]=time_span(user['user_id'][i])
```
