```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
import operator
```

```python
import_file = 'reviews3500.json'
df = pd.read_json(import_file, lines=True)
```

```python
df
```

```python
df['text']=df['text'].str.replace('"',' ')
```

```python
def createDictionary(text):
    dict = {}

    for comment in text:
        words = comment.split()
        for word in words:
            if not word in dict:
                dict[word]=1
            else:
                dict[word]=dict[word]+1
                
    stopset = stopwords.words('english')
    # Remove stopwords
    for stopword in stopset:
        if stopword in dict and stopword != 'not' and stopword != 'no':
            del dict[stopword]

    # Create list of sorted dictionary, sorted_cd, capped at 60 items
    sorted_cd = sorted(dict.items(), key=operator.itemgetter(1), reverse=True)
    sorted_cd = sorted_cd[:60]
    
    return sorted_cd
```

```python
def reviewsOfBusiness(s):
    texts = []
    for i in range(len(df['text'])):
        if df['business_id'][i] == s:
            texts.append(df['text'][i].replace('\n',' '))
    return texts
```
