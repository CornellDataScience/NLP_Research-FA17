```python
import pandas as pd
```

```python
import_file = 'dataset/review.json'
review = pd.read_json(import_file, lines=True)
```

```python
review
```

```python
reviewDates = review[['review_id','date']]
```

```python
reviewDates
```

```python
with open('reviewDates.json', 'w') as f:
    f.write(reviewDates.to_json(orient='records', lines=True))
```

```python
import_file = 'userReview.json'
userReview = pd.read_json(import_file, lines=True)
```

```python
rev = pd.merge(reviewDates, userReview, on='review_id', right_index=True)
```

```python
with open('userReviewDates.json', 'w') as f:
    f.write(rev.to_json(orient='records', lines=True))
```
