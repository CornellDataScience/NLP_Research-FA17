```python
import pandas as pd
```

```python
import_file = 'sampleDates.json'
date = pd.read_json(import_file, lines=True)
```

```python
date
```

```python
def time_span(user):
    user = [user]
    tmp = date.query('user_id == @user')['date']
    late = tmp.max()
    early = tmp.min()
    return abs((late - early).days)
```

```python
user = date['user_id'].drop_duplicates().reset_index()
del user['index']
user['time_span']=pd.Series()
user['time_sd']=pd.Series()
user
```

```python
for i in range(1000):
    user['time_span'][i]=time_span(user['user_id'][i])
```

```python
import matplotlib.pyplot as plt
user['time_span'].plot(kind='hist')
plt.show()
```

```python
from geopy.geocoders import Nominatim
areas = []
geolocator = Nominatim()
areas.append(geolocator.geocode("Las Vegas"))
areas.append(geolocator.geocode("Edinburgh"))
areas.append(geolocator.geocode("Stuttgart"))
areas.append(geolocator.geocode("Montreal"))
areas.append(geolocator.geocode("Toronto"))
areas.append(geolocator.geocode("Cleveland"))
areas.append(geolocator.geocode("Pittsburgh"))
areas.append(geolocator.geocode("Charlotte"))
areas.append(geolocator.geocode("Champaign"))
areas.append(geolocator.geocode("Madison"))
areas.append(geolocator.geocode("Phoenix"))
```

```python
areas
```

```python
area_names = ['Las Vegas','Edinburgh','Stuttgart','Montreal','Toronto','Cleveland','Pittsburgh','Charlotte','Champaign','Madison','Phoenix']
```

```python
def match_area(latitude, longitude):
    i = 0
    for area in areas:
        if abs(area.latitude-latitude)<0.6 and abs(area.longitude-longitude)<0.6:
            return area_names[i]
        i = i + 1
    return 'NaN'
```

```python
for i in range(1000):
    businessLoc['area'][i] = match_area(businessLoc['latitude'][i],businessLoc['longitude'][i])
```
